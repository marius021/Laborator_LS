```python
import pandas as pd

df = pd.read_csv("./iris.csv")

df.head()
```

```python
df["variety"].unique()
```

```python
df ["variety_encoded"] = df["variety"].replace({"Setosa": 0, "Versicolor": 1, "Virginica": 2})

df.head()
```

```python
print(len(df))
df =  df.dropna()

len(df)
```

```python
df = df.drop_duplicates()

len(df)
```

```python
df.describe()
```

```python
df.info()
```

```python
df['X'] = df[list(df.columns)[0:4]].values.tolist()

df
```

```python
df["Y"] = df["variety_encoded"].apply(lambda x: x + 1)

df
```

```python
df.describe()
```

```python
df.to_csv("iris_cleaned.csv", index=False)
```

# AI

```python
import pandas as pd

df = pd.read_csv("./iris_cleaned.csv")

df
```

```python
df["X"] = df["X"].apply(lambda x: list(x))
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)

X_train, y_train = train[[column for column in train.columns if column not in ["variety", "variety_encoded", "X", "Y"]]], train["Y"]
X_test, y_test = test[[column for column in test.columns if column not in ["variety", "variety_encoded", "X", "Y"]]], test["Y"]

knn = KNeighborsClassifier(n_neighbors=len(df["Y"].unique()))

knn.fit(X_train, y_train)

```

```python
from sklearn.metrics import accuracy_score
pred = knn.predict(X_test)

accuracy_score(y_test, pred)
```

```python
from pyod.models.knn import KNN

clf = KNN()

clf.fit(X_train, y_train)
```

```python
y_train_pred = clf.labels_

y_train_pred
```
